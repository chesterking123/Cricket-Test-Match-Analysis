# Filtering out the required columns

```python
data = df.filter(['Matchid', 'Players','Bat_Runs', 'BattingOrder', 'Bat_Bowler','Bat_Dismissal', '4s', '6s', 'Ball Faced', 'Dots','Maiden','extra','inning_bowl', 'Over', 'Wicket'], axis=1)
```

```python
data = data.fillna(0)
data.head()
```

```python
corr=data.corr()
sns.heatmap(corr, annot = True)
```

# UNIVARIATE PLOTTING

```python
(data['Bat_Dismissal'].value_counts()*100/len(data)).plot.bar()
plt.title('out type')
```

```python
(data['Wicket'].value_counts()*100/len(data)).plot.bar()
```

```python
data['Maiden'].value_counts().plot.bar()
```

```python
combined['player_of_match'].value_counts().head(10).plot.bar()
plt.title('Top 10 player of match')
```

```python
combined['winner'].value_counts().head(10).plot.bar()
plt.title('Top 10 teams wins')
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
data.plot(kind="box",subplots=True,layout=(4,4),sharex=False,sharey=False)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
data.plot(kind="density",subplots=True,layout=(4,4),sharex=False,sharey=False)
plt.show
```
