```python
import pandas as pd
import numpy as np 
from os import listdir
```

```python
match_info_final = pd.DataFrame(columns=['Team_1','Team_2', 'Gender', 'Start_Date','End_Date','Series','Match_No','Venue','City','Toss_Winner','Toss_Decision','Player_of_Match','Umpire_1','Umpire_2','Reverse_Umpire','TV_Umpire','Match_refree','Winner','Winner_Win_Type'])
match_info_final
```

```python
def match_info_generator(filename):
    csv_path = "E:\\BI\\test\\"
    df = open(csv_path+filename).read()
    Match_info = pd.DataFrame(columns=['Category','Result'])
    
    str_i="" 
    k=0
    rn=0
    c=1
    for i in range(14,len(df)):
        if(df[i]=='b' and df[i+1]=='a' and df[i+2]=='l' and df[i+3]=='l'):
            break
        if(df[i]==','):
            k+=1
            if(k!=1):
                Match_info.loc[rn,c-1:c]=str_i
                c+=1
            str_i=""
            continue
        if(df[i]=='\n'):
            Match_info.loc[rn,c-1:c]=str_i
            rn+=1
            c=1
            k=0
        str_i=str_i+df[i]    
    return (Match_info)
```

```python
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]

row_count=0
filenames = find_csv_filenames("E:\\BI\\test\\")
date1=""
date2=""
print("loading")
for filename in filenames:
    count=0
    date1=""
    date2=""
    d_index=0
    comp_alert = 0
    M=match_info_generator(filename)
    for i in range(len(M)):
        if(M['Category'][i]=='date' and count==0):
            count+=1
            date1=M['Result'][i]
        if(M['Category'][i]=='competition'):
            date2=M['Result'][i-1]
            d_index=i-1
            comp_alert = 1
        if(M['Category'][i]=='series'):
            date2=M['Result'][i-1]
            d_index=i-1
    for i in range(len(M)):
        if(i==3):
            match_info_final.loc[row_count,i:i+1]=date1
            continue
        if(i==4):
            match_info_final.loc[row_count,i:i+1]=date2
            break
        match_info_final.loc[row_count,i:i+1]=M['Result'][i] #for t1, t2 and gender
    cd=i+1
    
    
    if comp_alert == 0:
        for j in range(d_index+1,len(M)):
            match_info_final.loc[row_count,cd:cd+1]=M['Result'][j]
            cd+=1
    else:
        continue
                
    row_count+=1
    print(end=".")
match_info_final    
    
        
    
```
